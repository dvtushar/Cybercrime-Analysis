```python
# Load the dataset
file_path = 'new.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Select relevant columns
relevant_columns = [
    'Score',
    'Your Age Group', 
    'Studying or Working or Not Working', 
    'Do you know about Internet (Cyber) Crime?', 
    'Have you ever been victim of Cyber Crime?',
    'Did you loose any money?'
]

data = data[relevant_columns]

# Rename columns for clarity
data.rename(columns={
    'Score': 'Score',
    'Your Age Group': 'Age_Group',
    'Studying or Working or Not Working': 'Profession',
    'Do you know about Internet (Cyber) Crime?': 'CyberCrime_Aware',
    'Have you ever been victim of Cyber Crime?': 'Victim_Experience',
    'Did you loose any money?': 'Lost_Money'
}, inplace=True)

# print data
New_data = data
```
